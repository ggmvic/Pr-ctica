```python
'''
8. Crea un nuevo dataset con los pokemon Legendary
'''
legendary = df.loc[df['Legendary'] == True]
legendary
```

```python
'''
9. Calcula el promedio de stats de los pokemon Legendary (HP, Attack, Defense, Sp. Atk, Sp. Def, Speed) y los no Legendary
'''
df_legendary = legendary.describe().round(2)
df_legendary = df_legendary.drop(['count', 'std', 'min', '25%', '50%', '75%', 'max'], axis=0)
df_legendary = df_legendary.drop(['#', 'Total', 'Generation'], axis=1)
df_legendary.rename(index={'mean': 'Legendary'},inplace=True)
```

```python
df_no_legendary = df.loc[df['Legendary'] == False]
df_no_legendary = df_no_legendary.describe().round(2)
df_no_legendary = df_no_legendary.drop(['count', 'std', 'min', '25%', '50%', '75%', 'max'], axis=0)
df_no_legendary = df_no_legendary.drop(['#', 'Total', 'Generation'], axis=1)
df_no_legendary.rename(index={'mean': 'No Legendary'}, inplace=True)

```

```python
'''
10. Crea un nuevo dataframe con el resultado del anterior ejercicio comparando ambos promedios

ejemplo:
                HP  Attack  Defense  Sp. Atk  Sp. Def  Speed
Legendary       99      90       89       91       94     90
No Legendary    80      95       75       12       43     87

'''
df_average = pd.concat([df_legendary, df_no_legendary], axis=0)
df_average
```

```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']


# A. Una función que calcule el precio total de la compra
values = reduce(lambda a, b: a + b, list(map(lambda key: articulos[key], compra)))
print('Precio total: ', values, '€')

# B. Una función que calcule el precio total de la compra con un descuento del 10%

values_10_off = reduce(lambda a, b: a + b, list(map(lambda key: articulos[key]*0.9, compra)))
print('Precio con descuento: ', values_10_off, '€')

# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€
if values > 100:
    values_off_100 = values * 0.9
print(values_off_100, '€')

# D. Una función que calcule el precio total aplicando el IVA (21%)
values_iva = round((reduce(lambda a, b: a + b, list(map(lambda key: articulos[key], compra)))) * 1.21, 2)
print('Precio con IVA: ', values_iva, '€')
 
# E.lista los artículos cuyo precio es superior a 20€
sup_20 = list(filter(lambda key: articulos[key] > 20, compra))
print('Artículos con precio superior a 20: ', sup_20)
```

```python
'''
14. Dada una lista de tuplas con el nombre de un alumno, apellidos, curso y sus notas. 

 Define una funcion que reciba el curso y saque una lista en la que aparezca e nombre y apellidos y el promedio de sus notas.
 Puedes definir más funciones si lo consideras necesario.
'''

alumnos = [('Juan', 'Pérez', '1', [5, 6, 7, 8, 9]),
            ('Ana', 'García', '2', [5, 6, 7, 8, 9]),
            ('Luis', 'González', '1', [5, 6, 7, 8, 9]),
            ('María', 'Martínez', '2', [5, 6, 7, 8, 9]),
            ('Pedro', 'Rodríguez', '1', [5, 6, 7, 8, 9]),
            ('Lucía', 'Hernández', '2', [5, 6, 7, 8, 9]),
            ('Marta', 'López', '1', [5, 6, 7, 8, 9]),
            ('Sara', 'Díaz', '2', [5, 6, 7, 8, 9]),
            ('Javier', 'Sánchez', '1', [5, 6, 7, 8, 9]),
            ('Miguel', 'Fernández', '2', [5, 6, 7, 8, 9]),
            ('Sergio', 'Jiménez', '1', [5, 6, 7, 8, 9]),
            ('Sandra', 'Ruiz', '2', [5, 6, 7, 8, 9]),
            ('Pablo', 'Álvarez', '1', [5, 6, 7, 8, 9]),
            ('María', 'Gómez', '2', [5, 6, 7, 8, 9]),
]

promedio_notas = lambda notas: reduce(lambda x, y: x + y, notas) / len(notas)
crear_entrada = lambda alumnos: {'nombre': alumnos[0] + ' ' + alumnos[1], 'curso': alumnos[2], 'promedio': promedio_notas(alumnos[3])}
alumnos1 = list(map(crear_entrada,alumnos))
curso1 = list(filter(lambda alumnos1: alumnos1['curso'] == '1', alumnos1))
curso2 = list(filter(lambda alumnos1: alumnos1['curso'] == '2', alumnos1))
print('Curso 1:\n', curso1)
print('\n')
print('Curso 2:\n', curso2)

```
